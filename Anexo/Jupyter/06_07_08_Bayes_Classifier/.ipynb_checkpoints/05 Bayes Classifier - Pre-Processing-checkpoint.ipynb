# Generator Functions

When you have a function that need to read 50000 mails and then return them, it will be a lot of work if it  
has to return all of them at the same time.
To solve that, we will use GENERATOR FUNCTIONS:

> Instead of ***return*** we put ***yield*** (it "remembers the state where we had exited from")

```python
def generate_squares(N):
    for my_number in range(N):
        yield my_number ** 2
```

```python
# generate_squares(5) -> <generator object generate_squares at 0x00000274456402E0>
# To call a generate function we need something like this:

for i in generate_squares(5):
    print(i, end=' -> ')
```

## Email body extraction

```python
def email_body_generator(path):
    
    for root, dirnames, filenames in walk(path):
        # walk generates the file names in a directory by walking the tree from the top to the bottom
        # and it yields (not return) a tuple of three things: directory path (root),  
        # directory names (dirnames) and the file names (filenames)
        #
        # root: the path to our spam folder
        # dirnames: the subdirectories (we are not gonna use)
        # filenames: This is gonna be a list of names of all the files in our directory 
        for file_name in filenames:
            
            filepath = join(root, file_name)
            
            stream = open(filepath, encoding='latin-1')

            is_body = False
            lines = []

            for line in stream:
                if is_body:
                    lines.append(line)
                elif line == '\n':
                    is_body = True

            stream.close()

            email_body = '\n'.join(lines)
            
            yield file_name, email_body
```

```python
#In order to call the function above (that has yield) we need another function:
#This creates a DataFrame with the mails of the folders

def df_from_directory(path, classification):
    #classification is whether the email folder is gonna contain spam emails or legitimate emails 
    rows = []
    row_names = []
    
    for file_name, email_body in email_body_generator(path):
        rows.append({'MESSAGE': email_body, 'CATEGORY': classification})
        row_names.append(file_name)
        
    return pd.DataFrame(rows, index=row_names)
```

## Word Stems and Stemming

### Stemming
> Stemming is the process of reducing words to their base or root form.  
> The idea is to treat inflected or derived words in the same way.  
>
> For example: fisher, fishlike, fishing are reduced by the stemmer into fish.  
> Note that you may be ending with not a real word.
>
> We can use the PorterStemmer for english, but there are other stemmers like Snowball that let you choose the language.

```python
msg = 'All work and no play makes Jack a dull boy. To be or not to be. \
      Nobody expects the Spanish Inquisition!'
words = word_tokenize(msg.lower())

# stemmer = PorterStemmer()
stemmer = SnowballStemmer('english')


filtered_words = []
for word in words:
    if word not in stop_words:
        stemmed_word = stemmer.stem(word)
        filtered_words.append(stemmed_word)

print(filtered_words)
```

## Functions for Email Processing

```python
def clean_message(message, stemmer=PorterStemmer(), 
                 stop_words=set(stopwords.words('english'))):
    
    # Converts to Lower Case and splits up the words
    words = word_tokenize(message.lower())
    
    filtered_words = []
    
    for word in words:
        # Removes the stop words and punctuation
        if word not in stop_words and word.isalpha():
            filtered_words.append(stemmer.stem(word))
    
    return filtered_words
```

```python
clean_message(email_body)
```

```python
# Challenge: Modify function to remove HTML tags. Then test on Email with DOC_ID 2. 

def clean_message_no_html(message, stemmer=PorterStemmer(), 
                 stop_words=set(stopwords.words('english'))):
    
    # Remove HTML tags
    soup = BeautifulSoup(message, 'html.parser')
    text = soup.get_text()
    
    # Converts to Lower Case and splits up the words
    words = word_tokenize(text.lower())
    
    filtered_words = []
    
    for word in words:
        # Removes the stop words and punctuation
        if word not in stop_words and word.isalpha():
            filtered_words.append(stemmer.stem(word))
            #filtered_words.append(words)
    
    return filtered_words
```

### Create a Sparse Matrix for the Training Data

```python
#  To know what word is in a cell we just call column with the index of the row:
#vocab.VOCAB_WORD[1]
#vocab['VOCAB_WORD'][1]
#  But what if we know the word and we want to know the word ID (that in this case is the index)?
#  Then you have to create an INDEX for the column: (next cell)
vocab.head() #The head of the 2500 most frequents words in the emails, ordered by the frequency
```

```python
# Turn a particular column of a DataFrame into an Index

word_index = pd.Index(vocab.VOCAB_WORD)
#type(word_index) -> pandas.core.indexes.base.Index
#type(word_index[3]) -> str
```

```python
word_index.get_loc('origin') 
# This will tell you the WORD_ID, that is the index of the frequency of that word (0 is the most frequent)
# If it does not have an index that means that it is not in the 2500 most frequent words
```

```python
# Now, given the X_train matrix (that is a dataFrame), we are gonna create our sparse matrix.
# The columns of X_train are:
#   Index(DOC_ID)  0       1       2     ...    7668   7669   
# One first row of the X_train is for ex:
#   4885          origin  tri   instal   ...    None   None
# The columns of sparse matrix are:
#   DOC_ID  WORD_ID  LABEL  OCCURENCE
# Wih the row before it would be completed as:
#         Note: 
#              X_train.index[0] = 4885, 
#              y_train[4885] = 0 (The category, ham/spam, here, the LABEL),
#              word_index.get_loc('origin') = 274
#              word_index.get_loc('tri') = error (It is not one of the 2500 most frequent words)
#              word_index.get_loc('instal') = 328 
#   DOC_ID  WORD_ID  LABEL  OCCURENCE
#   4885     274      0        1
#   4885     328      0        1     
# -> As the word 'tri' wasn't in the most frequent, we didn't put it in our sparse matrix, 
# instead, we go to the next word.
# and so on.
```

```python
def make_sparse_matrix(df, indexed_words, labels):
    """
    Returns sparse matrix as dataframe.
    
    df: A dataframe with words in the columns with a document id as an index (X_train or X_test)
    indexed_words: index of words ordered by word id
    labels: category as a series (y_train or y_test)
    """
    
    nr_rows = df.shape[0]
    nr_cols = df.shape[1]
    # As we are gonna do a lot of lookups in our loop it is better to use a set:
    #(ex: if the words of our dataframe are part of the vocabulary list)
    word_set = set(indexed_words)
    dict_list = []
    
    for i in range(nr_rows):
        for j in range(nr_cols):
            
            word = df.iat[i, j]
            if word in word_set:
                doc_id = df.index[i]
                word_id = indexed_words.get_loc(word)
                category = labels.at[doc_id]
                
                item = {'LABEL': category, 'DOC_ID': doc_id,
                       'OCCURENCE': 1, 'WORD_ID': word_id}
                # Occurence = 1 as it is the first time we pass, and every time we discover
                #  that is part of our vocabulary, we'll add it to our dataframe
                #  We will change it later
                dict_list.append(item)
    # Create a DataFrame out of a Dictionary:
    return pd.DataFrame(dict_list)
```

```python
train_grouped = sparse_train_df.groupby(['DOC_ID', 'WORD_ID', 'LABEL']).sum()
train_grouped.head()
# We can see that the word 2 in DOC_ID 0 occurs once, and so on  
#vocab.at[2, 'VOCAB_WORD']
#data.MESSAGE[0]
```

```python
# We want to appear the DOC_ID in every row:

train_grouped = train_grouped.reset_index()
train_grouped.head()
```
