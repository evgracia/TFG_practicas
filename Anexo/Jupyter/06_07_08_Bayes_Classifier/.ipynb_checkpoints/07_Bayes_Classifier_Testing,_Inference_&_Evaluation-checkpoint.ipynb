## Set the Prior

> ***Prior*** is a guess or a belief about some quantity.  
> In our case we are gonna express a belief about the percentage of spam messages in the test dataset for clarity.  
> We use the Bayes formula:

$$P(Spam \, | \, X) = \frac{P(X \, | \, Spam) \, P(Spam)} {P(X)}$$

````python
PROB_SPAM = 0.3116
#We calculated it in our previous notebook
````

**Challenge**: Do you recall how to take the log of a numpy array? Calculate the log probabilities of the tokens given that the email was spam. This was stored in prob_token_spam. 

````python
np.log(prob_token_spam)
````

## Joint probability in log format

````python
# Probabilities that a email is spam given that it has the tokens that it has
joint_log_spam = X_test.dot(np.log(prob_token_spam) - np.log(prob_all_tokens)) + np.log(PROB_SPAM)
````

````python
joint_log_spam[:5]
````

````python
# Challenge:
# Calculate the log probability that the emails are non-spam given their tokens.
# Store the result in a variable called joint_log_ham
````

$$P(Ham \, | \, X) = \frac{P(X \, | \, Ham) \, (1-P(Spam))} {P(X)}$$

````python
joint_log_ham = X_test.dot(np.log(prob_token_ham) - np.log(prob_all_tokens)) + np.log(1-PROB_SPAM)
````

````python
joint_log_ham[:5]
````

````python
joint_log_ham.size
````

# Making Predictions

### Checking for the higher joint probability

$$P(Spam \, | \, X) \, > \, P(Ham \, | \, X)$$
<br>
<center> OR </center>
<br>
$$P(Spam \, | \, X) \, < \, P(Ham \, | \, X)$$

**Challenge**:  
Can you create the vector of predictions, our $\hat y$ ?  
Remember that spam emails should have the value 1 (true) and non-spam emails should have the value 0 (false).  
Store your results in a variable called `prediction`.

````python
prediction = joint_log_spam > joint_log_ham
````

````python
prediction[-5:]*1 
# To transform True/False to 0/1 we use *1
````

````python
y_test[-5:]
````

### Simplify the calculation

$$P(X \, | \, Spam) \, P(Spam) â‰   \frac{P(X \, | \, Spam) \, P(Spam)}{P(X)}$$

````python
joint_log_spam = X_test.dot(np.log(prob_token_spam)) + np.log(PROB_SPAM)
joint_log_ham = X_test.dot(np.log(prob_token_ham)) + np.log(1-PROB_SPAM)
````

# Metrics and Evaluation

## 1. Accuracy

> In Machine Learning:
> $$ accuracy = \frac{Nr. \, of \, correct \, predictions}{Total \, nr. \, of \, predictions}$$  
>
> The problem is when there is considerately fewer cases of spam/cancer cells/... than the normal ones.   
> Example:  
> If our model says that none of the persons in the country has cancer, it will have an accuracy of 96%, but it doesn't work.


````python
correct_docs = (y_test == prediction).sum()
print('Docs classified correctly', correct_docs)
numdocs_wrong = X_test.shape[0] - correct_docs
print('Docs classified incorrectly', numdocs_wrong)
````

````python
# Accuracy
correct_docs/len(X_test)
````

````python
fraction_wrong = numdocs_wrong/len(X_test)
print('Fraction classified incorrectly is {:.2%}'.format(fraction_wrong))
print('Accuracy of the model is {:.2%}'.format(1-fraction_wrong))
````

````python
true_pos = (y_test == 1) & (prediction == 1)
#Bitwise and -> & realiza la operaciÃ³n booleana a cada elemento
#&& -> realiza la operaciÃ³n and sobre los resultados finales
````

````python
true_pos.sum()
````

````python
# Challenge: 
# Create a numpy array that measures the False Positives for each datapoint. 
# Call this variable ```false_pos```. Then work out how many false positives there were. 
# After you've done that, do the same for the false negatives. 
# Store those in a variable called ```false_neg```
````

````python
false_pos = (y_test == 0) & (prediction == 1)
false_pos.sum()
````

````python
false_neg = (y_test == 1) & (prediction == 0)
false_neg.sum()
````

## 2. Recall (or Sensitivity) Score

### $$ Recall Score = \frac{True \, Positives}{True \, Positives \,+\, False \, Negatives} $$

> True positive in our case : When the email is spam and classified as spam
> False negative in our case : When the email is said to be not spam, but it is spam.  
> - If falses negatives = 0 -> Recall Score = 1  
> - If falses negatives > 0 -> Recall Score < 1  
>
> In the case of the cancer we talked before, this score would be 0, so now we can see that it is not reliable.  
> The problem is when you label all the emails as spam/cancer.


````python
# Challenge:
# Calculate the recall score. 
# Store it in a variable called ```recall_score```. 
# Print the value of the recall score as a percentage rounded to two decimal places.
````

````python
recall_score = true_pos.sum() / (true_pos.sum() + false_neg.sum())
print('Recall score is {:.2%}'.format(recall_score))
````

## 3. Precision Score (or Positive predictive value)

### $$ Precision Score = \frac{True \, Positives}{True \, Positives \,+\, False \, Positives} $$

> False positive in our case : When the email is said to be spam, but it is not spam.  
> So this is the number of correct emails labeled as spam / all the emails labeled as spam.  
> - If falses positives = 0 -> Precision Score = 1  
> - If falses positives > 0 -> Precision Score < 1  
>
> In the case of the cancer we talked before, this score would be 1, so it does not help.  
> However, if the model was that everyone is said to have cancer, this would be the metric that wouls show us the uselessness of our model.

````python
# Challenge:
# Calculate the precision of our naive bayes model. 
# Store the result in a variable called ```precision_score```.
# Print out the precision as a decimal number rounded to three decimal places. 
````

````python
precision_score = true_pos.sum() / (true_pos.sum() + false_pos.sum())
print('Precision score is {:.3}'.format(precision_score))
````

## F-Score or F1 Score

> This is the combination of the precision and Recall
> $$ F-Score = 2 * \frac{Precision \, * \, Recall}{Precision \, + \, Recall}$$
> So if we want to create a balanced classifier we try to maximize the F-Score
> This score is also between 0 and 1

````python
f1_score = 2 * (precision_score * recall_score) / (precision_score + recall_score)
print('F Score is {:.2}'.format(f1_score))
````
