```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

# Most popular machine learning module in Python:
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
X = DataFrame(data, columns = ['production_budget_usd'])
y = DataFrame(data, columns = ['worldwide_gross_usd'])
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X, y, alpha =0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $ // FEATURE, indpt vble')
plt.ylabel('Wordwide Gross $ (Revenue) // TARGET')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

We are going to draw the LINEAR REGRESION,
that is the straigth line:

$$ h_\theta(x) = \theta_0 + \theta_1 x $$

this is:
>$y = mx + c$

where: 
>c = intercept $(\theta_0)$
> 
> m = slope $(\theta_1)$ (more slope $\rightarrow$ more relationship between the variables)
> 
>$h_\theta(x^i)$: the fitted or predicted value of the element i
>
>$y^i$: the actual value of the element i

residual = $y^i - h(x^i)$

we try to min the the RESIDUAL SUM OF SQUARES: 

$$ \sum{(y^i - h(x^i))^2} $$

Now, we are going to run the LINEAR REGRESSION for calculating the slope and the intercept:

```python
#Creamos un objeto y lo almacenamos en regression:
regression = LinearRegression()
# X = names of the features, independet vbls
# y = name of our labels, target
# .fit(X,y) to run the linear regression
regression.fit(X,y)
```

and we are done, now we have to look at the results:

```python
# theta_1, Slope Cofficient
regression.coef_
print('theta_1 = ', regression.coef_)
# Intercept, theta_0
regression.intercept_
print('theta_0 = ', regression.intercept_)
```

- $\theta_1$ means there is a positive relationship between budget and revenue.
For each dolar we spent on producing the movie we should get around 3.11 $ in revenue in return.

- $\theta_0$ means that a movie with a budget of 0 losses 
$7e6$ approximately, and that makes no sense, it is no realistic

```python
plt.figure(figsize = (10,6))
plt.scatter(X, y, alpha =0.3, color = '#4a86e8')
plt.plot(X, regression.predict(X), color='#000000', linewidth=2)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $ // FEATURE, indpt vble')
plt.ylabel('Wordwide Gross $ (Revenue) // TARGET')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

We have seen that this model is not quite realistic, specially in the extremes situations, so we shouldn't believe too much in our model.

To predict revenue of a film whith a budget of 50 000 we just substitute in the ecuation:
$$h_\theta(50000) = \theta_0 + \theta_1 * 50000$$

How can we meassure how good our model is?

Answer: with the GOODNESS OF FIT ($r^2$ or $R^2$)


```python
regression.score(X, y)
```

this is the amount of variation in film revenue that is explained by the film's budget

as it is 0.55, it is really good, becauses thinking in a simple way, this model with a single feature (prod. budget) can explain around 55 percent of the variation that we see in worldwide movie earnings.

But we could use more features, or maybe use a non linear regression, that is what we are gonna see in the future
